# Up and Down the Python Data and Web Visualization Stack

USGS dataset listing every wind turbine in the United States:

```python
import pandas as pd
import vincent

vincent.initialize_notebook()
df = pd.read_csv('USGS_WindTurbine_201307_cleaned.csv')
df.head()
```

```python
mf_grouped = df.groupby('Turbine Manufacturer')
mean_rd = mf_grouped.mean().dropna().sort('Rotor Diameter')['Rotor Diameter']
rotor_diam = vincent.Bar(mean_rd)
rotor_diam.axis_titles(x='Turbine Manufacturer', y='Rotor Diameter')
rotor_diam.x_axis_properties(title_size=20, label_angle=50)
rotor_diam.y_axis_properties(title_size=20)
rotor_diam.axes[0].properties.labels.dx = vincent.values.ValueRef(value=5)
rotor_diam.axes[0].properties.labels.align = vincent.values.ValueRef(value='left')
rotor_diam.axes[0].properties.title.dy = vincent.values.ValueRef(value=115)
rotor_diam.scales[2].range = ['#6c71c4']
rotor_diam
```

```python
turbine_ct = mf_grouped.count().dropna().sort('Unique ID')['Unique ID']
turbines_vs_diam= pd.concat([turbine_ct, mean_rd], axis=1)
turbines_vs_diam.columns=['Turbine Count', 'Mean Rotor Diameter']
turbines_vs_diam = turbines_vs_diam.sort('Turbine Count', ascending=False)
turbines_vs_diam = turbines_vs_diam.drop(['unknown'])
turbines_vs_diam = turbines_vs_diam.dropna()
turbines_vs_diam
stacked = vincent.StackedBar(turbines_vs_diam)
stacked
```
